```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
data = pd.read_csv("Data-cleaning-for-beginners-using-pandas.csv")
```

```python
data.head(20)
```

```python
data.info()
```

```python
###For Easy_Apply
data.rename(columns = {'Easy Apply':'Easy_Apply'},inplace = True)
data["Easy_Apply"]=data["Easy_Apply"].str.replace("-1","False")
data
```

```python
avg_age=data['Age']. mean()
avg_age
data.Age.fillna(avg_age)
data['Age']=data.Age.fillna(avg_age)
data['Age']=data.Age.astype(int)
data
```

```python
### For Salary column

data['Salary']=data.Salary.str.replace("$"," ")

data['Salary']=data.Salary.str.replace("k", "000")

data[['Min_Salary', 'Max Salary']] = data['Salary'].str.split('-', expand=True)
data = data.drop(columns=['Salary'])
data
```

```python
### For Rating Column
data["Rating"] = data["Rating"].replace(-1.0, np.nan)
avg_rating=data['Rating'].mean()
data['Rating']=data.Rating.fillna(avg_rating)
data['Rating']=data.Rating.round(decimals=1)
data
```

```python
###For Location Column
data[['Location', 'Country_initials' ]] = data['Location'].str.split(',', expand=True)
data['Location']=data. Location.str.replace(" Aus"," ")
data['Country_initials']=data.Country_initials.str.replace("None", "Aus")
data[ 'Country_initials']=data.Country_initials.fillna('Aus')
data
```

```python
###For Established -Column
data["Established"]=data["Established"].astype(str)
data["Established"] = data["Established"].str.replace('-1','Unknown')
data
```
